```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

lifesat = pd.read_csv("dt.csv")
x = lifesat[["old"]].values
y = lifesat[["price"]].values

lifesat.plot(kind="scatter", grid=True, x="old", y="price")
plt.axis([15, 25, 0, 500000])
plt.show()

model = LinearRegression()

model.fit(x, y)

X_new = [[8]]
print(model.predict(X_new))
```
