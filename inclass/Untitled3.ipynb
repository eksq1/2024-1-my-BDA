```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

lifesat = pd.read_csv("data.csv")
x = lifesat[["dist"]].values
y = lifesat[["won"]].values

lifesat.plot(kind="scatter", grid=True, x="dist", y="won")
plt.axis([0, 12, 0, 35000])
plt.show()

model = LinearRegression()

model.fit(x, y)

X_new = [[8]]
print(model.predict(X_new))
```
